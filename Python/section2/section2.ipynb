```python
"""
Python if Statement Tutorial

The if statement is a fundamental control structure that allows conditional execution of code blocks.

Basic Structure:
if condition:
    # code to execute if condition is True

Example:"""
x = 10
if x > 5:
    print("x is greater than 5")

"""
1. Simple if Statement
- Evaluates a Boolean expression
- Executes code block only if condition is True"""
print("\nSimple if example:")
num = 5
if num > 0:
    print(f"{num} is a positive number")

"""
2. if-else Statement
- Provides alternative execution path"""
print("\nif-else example:")
temperature = 25
if temperature > 30:
    print("It's hot outside")
else:
    print("It's not hot outside")

"""
3. if-elif-else Chain
- Handles multiple conditions
- Evaluates from top to bottom"""
print("\nif-elif-else example:")
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
print(f"Score {score} = Grade {grade}")

"""
4. Nested if Statements
- if statements inside other if blocks"""
print("\nNested if example:")
age = 25
height = 175
if age >= 18:
    if height >= 160:
        print("You can ride this attraction")
    else:
        print("Minimum height not met")
else:
    print("Minimum age not met")

"""
5. Logical Operators
- Combine conditions using and/or/not"""
print("\nLogical operators example:")
user = "admin"
logged_in = True
if user == "admin" and logged_in:
    print("Access granted")
    
"""
6. Truth Value Testing
- Python evaluates non-Boolean objects in conditions"""
print("\nTruth value testing example:")
name = ""
if not name:
    print("Name is empty")

valid_numbers = []
if valid_numbers:
    print("List has numbers")
else:
    print("List is empty")

"""
7. Shorthand if-else (Ternary Operator)
- Conditional expressions in single line"""
print("\nTernary operator example:")
max_value = x if x > y else y
print(f"Maximum value is {max_value}")

"""
Common Pitfalls:
1. Using assignment (=) instead of comparison (==)
2. Forgetting colons at end of condition lines
3. Incorrect indentation
4. Overlooking falsy values (0, "", [], None, False)

Key Takeaways:
- Use indentation to define code blocks
- elif is Python's version of "else if"
- Conditions can be any expression evaluating to Boolean
- Empty collections and zero values are considered False
- Use parentheses for complex condition grouping
"""
```
